```python
from bs4 import BeautifulSoup 
```

```python
import requests
```

```python
url="https://www.flipkart.com/search?q=books+for+children&as=on&as-show=on&otracker=AS_Query_OrganicAutoSuggest_6_6_na_na_na&otracker1=AS_Query_OrganicAutoSuggest_6_6_na_na_na&as-pos=6&as-type=RECENT&suggestionId=books+for+children&requestId=eaf0e0fe-ab2c-4a12-b9f7-001eda10cd33&as-searchtext=books%20"
```

```python
url
```

```python
WebAllow=requests.get(url)
```

```python
WebAllow
```

```python
soup=BeautifulSoup(WebAllow.text)
```

```python
soup
```

```python
Books=soup.select(".s1Q9rs")
```

```python
Books
```

```python
bookname=[]
```

```python
for b in Books:
    bookname.append(b.text)
```

```python
bookname
```

```python
Price=soup.select("._8VNy32 ._30jeq3")
```

```python
Price
```

```python
bookprice=[]
```

```python
for p in Price:
    bookprice.append(p.text)
```

```python
bookprice
```

```python
Discount=soup.select("._8VNy32 span")
```

```python
Discount
```

```python
bookdiscount=[]
```

```python
for ds in Discount:
    bookdiscount.append(ds.text)
```

```python
bookdiscount
```

```python
Ratings=soup.select("._1lRcqv ._3LWZlK")
```

```python
Ratings
```

```python
rate=[]
```

```python
for r in Ratings:
    rate.append(r.text)
```

```python
rate
```

```python
Specifications=soup.select("._3Djpdu")
```

```python
Specifications
```

```python
speci=[]
```

```python
for s in Specifications:
    speci.append(s.text)
```

```python
speci
```

```python
OriginalPrice=soup.select("._3I9_wc")
```

```python
OriginalPrice
```

```python
oprice=[]
```

```python
for op in OriginalPrice:
    oprice.append(op.text)
```

```python
oprice
```
